```python
# Dataset has three column with null values, I have crossed with outliers with mean and median.
#Credit score null was replaced with mean value
from sklearn.preprocessing import Imputer
imputer=Imputer(missing_values='NaN',strategy='mean',axis=0) 
imputer.fit(dataset[["credit_score"]])
dataset["credit_scores"]=imputer.transform(dataset[["credit_score"]])
```

```python
dataset
```

```python
#Annual_mileage was replaced with mean, no much difference between the mean and median
from sklearn.preprocessing import Imputer
imputer=Imputer(missing_values='NaN',strategy='mean',axis=0) 
imputer.fit(dataset[["annual_mileage"]])
dataset["annual_mileages"]=imputer.transform(dataset[["annual_mileage"]])
```

```python
#monthly_mileage was replaced with median, no much difference between the mean and median
from sklearn.preprocessing import Imputer
imputer=Imputer(missing_values='NaN',strategy='mean',axis=0) 
imputer.fit(dataset[["monthly_mileage"]])
dataset["monthly_mileages"]=imputer.transform(dataset[["monthly_mileage"]])
```

```python
#Drop out the unwanted columns for the same created the new column
dataset=dataset.drop(["credit_score","monthly_mileage","annual_mileage"],axis=1)
```
